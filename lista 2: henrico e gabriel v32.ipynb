```python
#1 – Em um país fictício, o direito ao voto começa aos 16 anos. Crie um programa que receba a idade de uma pessoa e determine, usando apenas uma condição simples if, se ela já pode votar.
idade = int(input("Idade: "))

if idade >= 16:
    print("Você já pode votar.")
```

```python
#2 – Uma escola deseja automatizar a verificação do desempenho dos alunos. Leia duas notas, calcule a média e utilize uma estrutura if…else para informar se o aluno está aprovado (média ≥ 7) ou reprovado.
n1 = float(input("Nota 1: "))
n2 = float(input("Nota 2: "))

media = (n1 + n2) / 2

if media >= 7:
    print("Aprovado!")
else:
    print("Reprovado!")
```

```python
#3 – Um sistema meteorológico quer classificar a sensação térmica do dia. A partir de uma temperatura informada, determine se está fria, agradável, quente ou muito quente usando if…elif…else.
temp = float(input("Temperatura: "))

if temp < 15:
    print("Fria")
elif temp < 25:
    print("Agradável")
elif temp < 35:
    print("Quente")
else:
    print("Muito quente")
```

```python
#5 – Uma empresa deseja reajustar salários com base em faixas de rendimento. Receba o salário atual e aplique um aumento conforme as faixas (10%, 7% ou 5%), usando condicionais encadeadas. Exiba o novo salário.
sal = float(input("Salário: "))

if sal <= 1500:
    novo = sal * 1.10
elif sal <= 3000:
    novo = sal * 1.07
else:
    novo = sal * 1.05

print("Novo salário:", novo)
```

```python
#6 – Um aplicativo financeiro precisa identificar qual de três valores é o maior. Leia três números e mostre o maior deles utilizando estruturas condicionais.
a = float(input("A: "))
b = float(input("B: "))
c = float(input("C: "))

maior = a
if b > maior:
    maior = b
if c > maior:
    maior = c

print("Maior:", maior)
```

```python
#7 – Você está desenvolvendo um calendário digital. Crie um programa que receba um ano e determine se ele é bissexto.
ano = int(input("Ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or ano % 400 == 0:
    print("Bissexto")
else:
    print("Não bissexto")
```

```python
#10 – Um aplicativo exibe mensagens conforme o período do dia. A partir do horário informado (0 a 23), apresente a saudação apropriada.
h = int(input("Hora (0-23): "))

if 5 <= h < 12:
    print("Bom dia!")
elif 12 <= h < 18:
    print("Boa tarde!")
elif 18 <= h < 24:
    print("Boa noite!")
else:
    print("Boa madrugada!")
```

```python
#11 – Uma calculadora avançada deve permitir que o usuário some quantos números quiser. Peça números repetidamente até que o valor 0 seja digitado. Ao final, mostre a soma total.
soma = 0

while True:
    n = float(input("Número (0 para sair): "))
    if n == 0:
        break
    soma += n

print("Soma total =", soma)
```

```python
#13 – Uma máquina de simulação deve somar números sequenciais começando em 1 até que a soma ultrapasse 100, mostrando cada número utilizado. Utilize um laço while.
soma = 0
n = 1

while soma <= 100:
    print("Somando:", n)
    soma += n
    n += 1

print("Soma final:", soma)
```

```python
#14 – Uma pesquisa está registrando idades. Continue lendo idades até que o usuário digite um valor negativo. Ao final, calcule e exiba a média das idades digitadas.
idades = []
while True:
    idade = int(input("Idade: "))
    if idade < 0:
        break
    idades.append(idade)

media = sum(idades) / len(idades)
print("Média:", media)
```

```python
#15 – Um cronômetro digital deve realizar uma contagem regressiva personalizada. Receba um número inicial e exiba a contagem até 0 usando um while.
n = int(input("Início: "))

while n >= 0:
    print(n)
    n -= 1
```

```python
#20 – Um sistema estatístico deve analisar números fornecidos pelo usuário. Leia 7 números e determine qual deles é o maior.
maior = None

for _ in range(7):
    n = float(input("Número: "))
    if maior is None or n > maior:
        maior = n

print("Maior:", maior)
```

```python
#23 – Um módulo de análise precisa classificar uma lista de valores. Leia 10 números e exiba quantos são positivos, negativos e quantos são iguais a zero.
pos = neg = zero = 0

for _ in range(10):
    n = float(input("Número: "))
    if n > 0:
        pos += 1
    elif n < 0:
        neg += 1
    else:
        zero += 1

print("Positivos:", pos)
print("Negativos:", neg)
print("Zeros:", zero)
```

```python
#27 – Crie um sistema de login com limite de tentativas. O usuário terá no máximo 3 tentativas para digitar a senha correta. Caso acerte antes, use break para encerrar o laço.
senha_correta = "python123"

for tent in range(3):
    senha = input("Senha: ")
    if senha == senha_correta:
        print("Acesso permitido.")
        break
else:
    print("Acesso bloqueado.")
```

```python
#28 – Desenvolva um programa que leia 10 números e some apenas os valores positivos. Caso um número negativo seja digitado, utilize continue para ignorá-lo.
soma = 0

for _ in range(10):
    n = float(input("Número: "))
    if n < 0:
        continue
    soma += n

print("Soma dos positivos:", soma)
```

```python
#29 – Simule um acumulador que percorre números até que a soma ultrapasse 200. Quando isso ocorrer, encerre o laço usando break.
soma = 0
n = 1

while True:
    soma += n
    if soma > 200:
        break
    n += 1

print("Soma final:", soma)
```

```python
#30 – Um programa de edição quer contar quantas palavras o usuário digitou. Continue lendo palavras até que o usuário digite “fim”. Depois, exiba quantas palavras válidas foram informadas.
cont = 0

while True:
    palavra = input("Palavra: ")
    if palavra.lower() == "fim":
        break
    cont += 1

print("Total de palavras digitadas:", cont)
```
