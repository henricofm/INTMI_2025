```python
#1. Crie uma lista com 15 números inteiros digitados pelo usuário. Em seguida, remova todos os números repetidos, mantendo a ordem original.

numeros = []
for _ in range(15):
    numeros.append(int(input("Digite um número: ")))

sem_repetidos = []
for n in numeros:
    if n not in sem_repetidos:
        sem_repetidos.append(n)

print(sem_repetidos)
```

```python

#2. Escreva um programa que leia 10 nomes e 10 idades. Armazene os nomes em uma lista e as idades em outra. Em seguida, exiba o nome da pessoa mais velha.
nomes = []
idades = []

for i in range(10):
    nome = input(f"Digite o nome da pessoa {i+1}: ")
    idade = int(input(f"Digite a idade da pessoa {i+1}: "))

    nomes.append(nome)
    idades.append(idade)

# encontra o índice da maior idade
mais_velho = idades.index(max(idades))

print("\nA pessoa mais velha é:", nomes[mais_velho])
print("Idade:", idades[mais_velho])


```

```python
#3. Faça um programa que leia uma lista de 10 números e divida todos os valores ímpares por 3 e os pares por 2, substituindo os valores originais.
numeros = []
for _ in range(10):
    numeros.append(int(input("Número: ")))

for i in range(len(numeros)):
    if numeros[i] % 2 == 0:
        numeros[i] /= 2
    else:
        numeros[i] /= 3

print(numeros)
```

```python
#5. Leia 10 números inteiros e armazene-os em uma lista. Em seguida, crie uma segunda lista apenas com os números primos da primeira.
def primo(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

lista = [int(input()) for _ in range(10)]
primos = [n for n in lista if primo(n)]
print(primos)


```

```python
#13. Solicite uma matriz 3x3 preenchida pelo usuário. Verifique se ela é uma matriz identidade.
mat = [[int(input()) for _ in range(3)] for _ in range(3)]

identidade = True
for i in range(3):
    for j in range(3):
        if (i==j and mat[i][j] != 1) or (i!=j and mat[i][j] != 0):
            identidade = False

print("Identidade" if identidade else "Não é identidade")
```

```python
#14. Faça um programa que leia uma matriz 3x3 e verifique se ela é simétrica (igual à sua transposta).
mat = [[int(input()) for _ in range(3)] for _ in range(3)]

simetrica = True
for i in range(3):
    for j in range(3):
        if mat[i][j] != mat[j][i]:
            simetrica = False

print("Simétrica" if simetrica else "Não simétrica")

```

```python
#17. Dada uma matriz 3x3 com valores reais, calcule o determinante (sem usar bibliotecas externas).
m = [[float(input()) for _ in range(3)] for _ in range(3)]

det = (m[0][0]*(m[1][1]*m[2][2] - m[1][2]*m[2][1]) -
       m[0][1]*(m[1][0]*m[2][2] - m[1][2]*m[2][0]) +
       m[0][2]*(m[1][0]*m[2][1] - m[1][1]*m[2][0]))

print(det)
```

```python
#18. Faça um programa que leia duas matrizes 3x3 e calcule a matriz soma e a matriz diferença.
A = [[int(input()) for _ in range(3)] for _ in range(3)]
B = [[int(input()) for _ in range(3)] for _ in range(3)]

soma = [[A[i][j] + B[i][j] for j in range(3)] for i in range(3)]
dif  = [[A[i][j] - B[i][j] for j in range(3)] for i in range(3)]

print("Soma:", soma)
print("Diferença:", dif)
```

```python
#25. Crie um dicionário que armazene o nome e o salário de 5 funcionários. Calcule o reajuste de 10% para salários inferiores a R$ 2.000 e atualize o dicionário.
func = {}

for _ in range(5):
    nome = input()
    salario = float(input())
    if salario < 2000:
        salario *= 1.10
    func[nome] = salario

print(func)

```

```python
#26. Faça um programa que armazene em um dicionário o nome de 5 alunos e uma lista com suas 3 notas. Depois, exiba o nome e a média de cada aluno.
alunos = {}

for _ in range(5):
    nome = input()
    notas = [float(input()) for _ in range(3)]
    alunos[nome] = sum(notas)/3

print(alunos)
```

```python
#28. Solicite ao usuário 10 palavras e armazene em um dicionário onde a chave é a primeira letra e o valor é uma lista de palavras que começam com essa letra.
d = {}

for _ in range(10):
    palavra = input()
    letra = palavra[0].lower()
    d.setdefault(letra, []).append(palavra)

print(d)
```

```python
#29. Crie um dicionário onde as chaves são nomes de produtos e os valores são listas com preços praticados em três lojas diferentes. Encontre o produto com maior variação de preço.
produtos = {
    "arroz": [10, 11, 12],
    "feijao": [8, 7, 9],
    "oleo": [6, 7, 8]
}

maior = None
nome = ""

for prod, precos in produtos.items():
    variacao = max(precos) - min(precos)
    if maior is None or variacao > maior:
        maior = variacao
        nome = prod

print("Maior variação:", nome)
```

```python
#30. Implemente um programa que leia uma frase e armazene em um dicionário o número de ocorrências de cada palavra (ignorando letras maiúsculas/minúsculas).
frase = input().lower().split()
cont = {}

for p in frase:
    cont[p] = cont.get(p, 0) + 1

print(cont)
```
